```julia
function f(z::Complex{Float64}, c::Complex{Float64})
    return (z^2 + c)
end
```

```julia
r_min = -2.2; r_max = .5
i_min = -1.2; i_max = 1.2 

size = 500
max_iters = 1000;
```

```julia
function make_cvals(r_min,i_min,r_max,i_max,size)
    c_vals = zeros(Complex{Float64}, (size,size))
    for i in 1:size
        for j in 1:size
            i_n = (i/size)*(r_max - r_min) + r_min
            j_n = (j/size)*(i_max - i_min) + i_min
            c_vals[i,j] = i_n + j_n*im
        end
    end
    return c_vals
end
```

```julia
function iter_until_escape(c, max_iters)
    fn = f(ComplexF64(0,0), c)
    iter = 1
    while (abs(fn) <= 2 && iter < max_iters) 
        fn = f(fn, c)
        iter+=1
    end
    
    return iter::Int64
end
    

function make_escape_map(c_vals,size)
    p = zeros(Float64, size,size)
    for i in 1:size
        for j in 1:size
            iter = iter_until_escape(c_vals[i,j], max_iters)
            p[i,j] = iter
        end
    end
    return p
end
```

```julia
@time c_vals = make_cvals(r_min, i_min, r_max, i_max, size);
```

```julia
@time e_map = make_escape_map(c_vals,size);
```
